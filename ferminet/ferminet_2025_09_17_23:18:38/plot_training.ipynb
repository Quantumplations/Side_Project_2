```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
step,energy,ewmean,ewvar,pmove = np.loadtxt('train_stats.csv', unpack=True, skiprows=1, delimiter=',')
with plt.style.context(['science']):
    plt.figure(figsize=(10,6))
    plt.plot(step, energy, label='Energy')
    plt.plot(step, ewmean, label='EW Mean')
    plt.plot(step, ewvar, label='EW Variance')
    plt.plot(step, pmove, label='P Move')
    plt.xlabel('Step')
    plt.ylabel('Values')
    plt.title('Training Statistics Over Steps')
    plt.legend()
    plt.grid()
    plt.show()
```
